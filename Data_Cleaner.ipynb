# Import

```python
import pandas as pd
import numpy as np

df = pd.read_csv('BVBRC_genome_amr.csv')

print(len(df))
print(df.isnull().sum())
```

# Delete Unused Columns

```python
df = df.drop(columns=['Measurement', 'Measurement Sign', 'Measurement Value', 'Measurement Unit', 'Laboratory Typing Method', 'Laboratory Typing Method Version', 'Laboratory Typing Platform', 'Vendor', 'Testing Standard', 'Testing Standard Year', 'Computational Method', 'Computational Method Version', 'Computational Method Performance', 'Source'])
print(df.isnull().sum())
```

# Drop Rows w/ Missing Cells

```python
df.dropna(inplace=True)
```

```python
print(df.isnull().sum())
print(len(df))
```

# Save To CSV

```python
df.to_csv('BVBRC_genome_amr_cleaned.csv', index=False)
```
